```python
subset=orders[['CustomerID','amount_spent']].groupby('CustomerID').sum()
```

```python
subset.reset_index(inplace=True)
subset
```

```python
subset_sorted=subset.amount_spent.sort_values(ascending=False)
subset_sorted
```

```python
subset_sorted.quantile([0.75, 0.95])
```

```python
orders['Customer Type']=''
orders.loc[orders['amount_spent']>5840.182, 'Customer Type'] = 'VIP'
orders.loc[orders['amount_spent']<1661.640, 'Customer Type'] = 'Normal'
orders.loc[(orders['amount_spent']<=5840.182) & (orders['amount_spent']>=1661.640), 'Customer Type'] = 'Preferred'
orders['Customer Type'].value_counts()
```
