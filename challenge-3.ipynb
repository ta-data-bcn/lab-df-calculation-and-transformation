## How to label customers whose aggregated `amount_spent` is in a given quantile range?



We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

```python
# your code here

customers = orders.groupby('CustomerID')
total_amount = customers.aggregate({'amount_spent': np.sum,})

total_amount
```

```python
orders["Quantiles"] = pd.qcut(orders["amount_spent"],
                                    q=[0, 0.25, 0.5, 0.75, 0.95, 1], 
                                    labels=["Very Low", "Low", "Medium", "Preferred", "VIP"])
```

```python
orders
```

```python
vip_customers = orders[orders["Quantiles"]== 'VIP']

vip_customers 
```

```python
preferred_customers = orders[orders["Quantiles"]== 'Preferred']

preferred_customers
```

## Q2: How to identify which country has the most VIP Customers?

```python
orders.groupby("Country")["Quantiles"].value_counts()
```

```python
pd.crosstab(orders['Country'],orders['Quantiles'])

# Netherlandsv has the most VIP Customers
```
